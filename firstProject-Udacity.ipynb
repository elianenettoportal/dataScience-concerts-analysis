```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
import seaborn as sns
import babel.numbers
import locale
%matplotlib inline
# import csv music dataset
df = pd.read_csv('./concert_dataset.csv')
df.head()
```

```python
# pandas shape() - it enables us to obtain the shape of a DataFrame (rows vs columns)
df.shape
```

```python
# pandas describe() - it returns the statistical summary of the dataframe or series
df.describe()
```

```python
# pandas - remove the column with null values
df2=df.dropna()
df2=df.dropna(axis=1)
print(df2)
```

```python
# Convert currency to decimal - remove $ and , remaining only the dot. Store it in a new variable
newTotal = df2["TOTAL"].str.replace('$', '').str.replace(',','')
# Convert to decimal
newTotal = pd.to_numeric(newTotal)
# Add it to a new column 
df2['New Total'] = newTotal
df2
```

```python
# Question 1 - What Venue sold more tickets

solded = df2.groupby('VENUE')['TICKETS_SOLD'].sum()
print(solded)
```

```python
# ANSWER 1 -> NAPA sold more tickets
```

```python
#Question 2 - What category made more revenue

# group by category and find the greater value
groupedByCategory = df2.groupby('ORIGINAL/COVER')['New Total'].sum()
print(groupedByCategory)
```

```python
# cover category sold more tickets 130,8015.00
```

```python
#Question 3 - Is capacity related to earning

# convert total to decimal
newTotalV = df2["TOTAL"].str.replace('$', '').str.replace(',','')
newTotalV = pd.to_numeric(newTotalV)
df2['New Total'] = newTotalV
df2
```

```python
# Group by Venue and Sum Tickets Sold
ticketsSold = df2.groupby('VENUE')['TICKETS_SOLD'].sum()
print(ticketsSold)
```

```python
# Group by Venue and sum total revenue
venueRevenue = df2.groupby('VENUE')['New Total'].sum()
print(venueRevenue)
```

```python
# Group by Venue and sum capacity
venueCapacity = df2.groupby('VENUE')['CAPACITY'].sum()
print(venueCapacity)
```
